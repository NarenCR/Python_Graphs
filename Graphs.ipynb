```python
x = np.linspace(-1,6,30)
x
```

```python
y = np.linspace(-1,6,30)
y
```

```python
z = x + y 
z
```

```python
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
z1 = sin_fun(x,y)
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z1)
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(df['sepal_length'],df['sepal_width'],df['petal_width'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_width')
```
